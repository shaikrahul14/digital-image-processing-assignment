```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

def create_discs(M, N, border, n, r1, r2, Vf=None, Vb=None):
    # Make a blank grayscale image with the required border and size.
    img = np.zeros((M+2*border, N+2*border), dtype=np.uint8)
    # White should be used as the backdrop colour if Vb is absent.
    if Vb is None:
        vb = 255
        img[:] = vb
    else:
        vb = np.random.choice(Vb)
        img[:] = vb
    # Try to distribute n non-overlapping discs at random.
    i = 0
    while i < n:
        # If we've tried too many times and still can't find a non-overlapping position, raise the image size and try again.
        for j in range(2*n):
            # Choose the disc's radius at random.
            r = np.random.randint(r1, r2+1)
            print('-------------\nradius : ', r)
            # Choose at random where the disc should be in the middle.
            cx = np.random.randint(border+r, M+border-r+1)
            cy = np.random.randint(border+r, N+border-r+1)
            print(cx, cy)
            # Verify to see if the disc overlaps any of the discs that were previously put.
            subimg = img[cy-r-1:cy+r+1, cx-r-1:cx+r+1]
            cv2_imshow(subimg)
            print('sum', np.sum(subimg), np.sum(255 - subimg))
            # check to see if there are any white pixels after negation. 
            if np.sum(vb - subimg) == 0:
                # Set the intensity of the disc to maximum if Vf is not provided.
                if Vf is None:
                    img = cv2.circle(img, (cx, cy), r, 0, -1)
                    cv2_imshow(img)
                # Unless otherwise specified, sample the intensity values from Vf and Vb.
                else:
                    vf = np.random.choice(Vf)
                    if vf == vb:
                      continue
                    # vb = np.random.choice(Vb)
                    print(vf, vb)
                    img = cv2.circle(img, (cx, cy), r, int(vf), -1)
                    img[img == 0] = vb
                    cv2_imshow(img)
                i += 1
                break
        else:
            # beginning fresh with double the height and width
            M *= 2
            N *= 2
            img = np.zeros((M+2*border, N+2*border), dtype=np.uint8)
            i = 0
            if Vb is None:
                img[:] = 255
            else:
                vb = np.random.choice(Vb)
                img[:] = vb
            cv2_imshow(img)
            
    # setting the image's border
    img[:border,:] = 0
    img[-border:,:] = 0
    img[:,0:border] = 0
    img[:,-border:] = 0
    return img
```

```python
import cv2
import numpy as np

# Example usage of create_discs function
Vf=[0, 64, 128, 192, 255]
Vb=[127, 129, 131, 133, 135, 137, 139, 141, 143, 145, 147, 149, 151, 153, 155, 157, 159, 161, 163, 165, 167, 169, 171, 173, 175, 177, 179, 181, 183, 185, 187, 189, 191, 193, 195, 197, 199, 201, 203, 205, 207, 209, 211, 213, 215, 217, 219, 221, 223, 225, 227, 229, 231, 233, 235, 237, 239, 241, 243, 245, 247, 249, 251, 253, 255]
img = create_discs(500, 500, 50, 70, 20, 35, Vf, Vb)
cv2_imshow(img)
cv2.imwrite('output.png', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

The code above is a Python function create_discs that generates a binary image containing n non-overlapping disks or circles of random radii within a border of size border pixels. The image has a size of M x N pixels.

The function takes the following parameters:

- M: an integer specifying the height of the image.
- N: an integer specifying the width of the image.
border: an integer specifying the size of the border around the image.
- n: an integer specifying the number of non-overlapping circles to be generated.
- r1: an integer specifying the minimum radius of the circles.
- r2: an integer specifying the maximum radius of the circles.
- Vf: a list of integers or None specifying the possible intensity values of the circles. If None, the circles will be black (intensity = 0).
- Vb: a list of integers or None specifying the possible intensity values of the background. If None, the background will be white (intensity = 255).

The function creates a grayscale image with the required size and border. If Vb is None, the background color is white (intensity = 255); otherwise, a random value is chosen from Vb. The function then attempts to distribute n non-overlapping circles at random in the image. For each circle, the function selects a random radius between r1 and r2, as well as a random location for the center of the circle within the image bounds. It then checks whether the circle overlaps with any of the previously placed circles by examining the corresponding sub-image. If there are no overlapping pixels, the function draws the circle on the image with a specified intensity. If Vf is None, the intensity of the circle is maximum (intensity = 0); otherwise, a random value is chosen from Vf. If vf equals vb, the function ignores the circle and continues to the next one.

If the function is unable to find a non-overlapping position for a circle after 2*n attempts, it increases the image size by doubling its height and width and tries again. The function returns the final binary image with the specified border.
